## saq 1_2 ~ 44_2까지의 결측치 값 처리 방법
- 위의 결측치 값들은 모두 동일한 형태로, 1번의 선행 질문에서 1번 (중요하지 않다)을 답한 사람들은 2번에 답하지 않는 구조이다. 
- 즉, 1번에서 1으로 응답했다면 2번은 null이어야하고, 1번에서 2~5로 응답하였으면 2번이 null이면 안된다. 
- 1번에서 1로 응답했지만 2번을 응답한 경우와, 1번에서 2~5로 응답했지만 2번을 응답하지 않은 경우를 찾아보자.

```python
for i in range(1, 45):
    if len(train_2020.loc[lambda x : (x['saq'+str(i)+'_2'].notnull()) & (x['saq'+str(i)+'_1'] == 1)]) != 0:
        print(i)
```
